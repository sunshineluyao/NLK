```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
df1 =pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/NLK/main/Data/inexperienced.csv")
df1
```

```python
fig1 = go.Figure()
fig1.add_trace(go.Scatter(mode='markers+lines',
    x=df1['lambda'],
    y=df1['MSE'],
    marker=dict(
            color='LightSkyBlue',
            size=6,
            line=dict(
                color='MediumPurple',
                width=6
            )
        ),
))
fig1.add_annotation(x=1, y=2.085,
            text='The minimum of MSE=2.085',
            showarrow=True,
            arrowhead=6)


fig1.update_xaxes(
        tickangle = 0,
        title_text = "$\lambda$",
        title_font = {"size": 20},
        title_standoff = 25)

fig1.update_yaxes(
        title_text = "MSE",
        title_standoff = 25)
fig1.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 0.05
    )
)
fig1.update_layout(showlegend=False)
fig1.update_layout(template="plotly_white", title="Mean Squared Error (MES): Inexperienced Bidders")

fig1.update_layout(font=dict(
        family="Times New Roman",
        size=26,
        color='#000001'
    ))
```
